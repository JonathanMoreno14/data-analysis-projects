# Data Cleaning

### Removing Outlier

```python
q1 = df['price'].quantile(0.25)
q3 = df['price'].quantile(0.75)
iqr = q3 - q1
```

```python
q1, q3, iqr
```

```python
upper_limit = q3 + (1.5 * iqr)
lower_limit = q1 - (1.5 * iqr)
lower_limit, upper_limit
```

```python
sns.boxplot(df['price'])
```

```python
# Finding the outliers
df.loc[(df['price']> upper_limit) | (df['price'] < lower_limit)]
```

```python
# Trimming - (Removing) the outlier data 
df2 = df.loc[(df['price'] < upper_limit) | (df['price'] > lower_limit)]
print('Before the removal of the outliers: ', len(df))
print('After removing the outliers: ', len(df2))
print('Outliers: ', len(df)-len(df2))
```

```python
sns.boxplot(df2['price'])
```

```python
# Capping - Changing the outlier values to upper or lower limit values
df2 = df.copy()
df2.loc[(df2['price'] > upper_limit),'price'] = upper_limit
df2.loc[(df2['price'] < lower_limit),'price'] = lower_limit
```

```python
# Boxplot with no outliers 
sns.boxplot(df2['price'])
```
